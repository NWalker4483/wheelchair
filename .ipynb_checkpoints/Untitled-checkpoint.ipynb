```python
from imutils.video import VideoStream, FileVideoStream
import imutils
import time
import cv2
vs = FileVideoStream("data/test.mp4").start()
from  matplotlib import pyplot as plt
%matplotlib inline
```

```python
frame = vs.read()
frame = imutils.resize(frame, width=200)
frame = frame[:,:,::-1] 
```

```python
gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
im2 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)
```

```python
import numpy as np 

# Creating kernel
kernel = np.ones((5, 5), np.uint8)
  
# Using cv2.erode() method 
image = cv2.erode(im2, kernel) 
plt.imshow(scan(gray))
```

```python
def scan(img):
    n = 3
    res = np.zeros_like(img)
    print(res.shape)
    for y in range(res.shape[0]):
        for x in range(res.shape[1]):
            valid = 0 
            for y_mod in range(-n,n + 1):
                for x_mod in range(-n,n + 1):
                    _x = x + x_mod
                    _y = y + y_mod
                    if _y < res.shape[0] and _x < res.shape[1]:
                        if img[_y][_x] > 20:
                            res[_y][_x] = 255
    return res
scan(gray)
```
